```python
fig=plt.figure(figsize=(20,20))
ax1=fig.add_subplot(3,2,1)
ax2=fig.add_subplot(3,2,2)
ax3=fig.add_subplot(3,2,3)
ax4=fig.add_subplot(3,2,4)
ax5=fig.add_subplot(3,2,5)
ax6=fig.add_subplot(3,2,6)


ax1.scatter(recent_grads.Sample_size,recent_grads.Median,color='red')
ax1.set_title("Sample_size & Median",fontweight='bold',fontsize=10)
ax1.set_xlabel("Sample_size",fontsize=10)
ax1.set_ylabel("Median",fontsize=10)

ax2.scatter(recent_grads.Sample_size,recent_grads.Unemployment_rate,color='green')
ax2.set_title("Sample_size & Unemployment_rate",fontweight='bold',fontsize=10)
ax2.set_xlabel("Sample_size",fontsize=10)
ax2.set_ylabel("Unemployment_rate",fontsize=10)

ax3.scatter(recent_grads.Full_time,recent_grads.Median,color='yellow')
ax3.set_title("Full_time & Median",fontweight='bold',fontsize=10)
ax3.set_xlabel("Full_time",fontsize=10)
ax3.set_ylabel("Median",fontsize=10)

ax4.scatter(recent_grads.ShareWomen,recent_grads.Unemployment_rate,color='purple')
ax4.set_title("ShareWomen & Median",fontweight='bold',fontsize=10)
ax4.set_xlabel("ShareWomen",fontsize=10)
ax4.set_ylabel("Median",fontsize=10)

ax5.scatter(recent_grads.Men,recent_grads.Median,color='blue')
ax5.set_title("Men & Median",fontweight='bold',fontsize=10)
ax5.set_xlabel("Men",fontsize=10)
ax5.set_ylabel("Median",fontsize=10)

ax6.scatter(recent_grads.Women,recent_grads.Median,color='orange')
ax6.set_title("Women & Median",fontweight='bold',fontsize=10)
ax6.set_xlabel("Women",fontsize=10)
ax6.set_ylabel("Median",fontsize=10)

plt.show()
```

```python
cols=['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig=plt.figure(figsize=(20,20))
for i in range(len(cols)):
    ax=fig.add_subplot(4,2,i+1)
    ax.hist(recent_grads[cols[i]])
plt.show()
```
